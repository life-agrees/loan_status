```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
%matplotlib inline
```

```python
loan = pd.read_csv('cleaned_loan.csv')
```

```python
loan.head(10)
```

```python
#converting some float value to integers. 
```

```python
loan['LoanAmount'] = loan['LoanAmount'].astype(int)
loan['Credit_History'] = loan['Credit_History'].astype(int)
loan['LoanTerm'] = loan['Loan_Amount_Term'] / 12

loan.drop(['Loan_Amount_Term'], axis=1, inplace=True)

label_encoder = LabelEncoder()

loan['LOAN_STATUS']= label_encoder.fit_transform(loan['Loan_Status'])

loan.drop(['Loan_Status'], axis=1, inplace=True)
```

```python
loan.head(10)
```

```python
df = ['Gender','Married','Education','Self_Employed','Property_Area','Dependents','ApplicantIncome','CoapplicantIncome','LoanAmount','Credit_History','LoanTerm','LOAN_STATUS']

loan = loan[df]

loan.head(10)
```

```python
label_encoder = {}
cat_col = ['Gender','Married','Education','Self_Employed','Property_Area','Dependents']
for cols in cat_col:
    label_encoder[cols] = LabelEncoder()
    loan[cols] = label_encoder[cols].fit_transform(loan[cols])
```

```python
num_cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','LoanTerm']
scale = StandardScaler()
loan[num_cols] = scale.fit_transform(loan[num_cols])
```

#let's drop some data the would be important for a mortgage loan
#by doing so we would have to check the features correlation

```python
corr_matrix = loan.corr()

plt.figure(figsize=(12,10))

sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')

plt.show()

correlation_target = corr_matrix['LOAN_STATUS'].sort_values(ascending=False)
print(correlation_target)
```

```python
loan.columns
```

```python


X = loan.drop(['LOAN_STATUS','Education','Self_Employed','Dependents'], axis=1)

y = loan['LOAN_STATUS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X.columns, X.shape
```

```python
def models():
    model = {
        'LogR': LogisticRegression(),
        'Ranf': RandomForestClassifier(n_estimators=150),
        'XGB': XGBClassifier()
    }
    return model
```

```python
data_model = models()

for name, model in data_model.items():
    model.fit(X_train, y_train)
    
```

```python
for name,model in data_model.items():
    predictions = model.predict(X_test)
    print(f'MODEL NAME:{name}')
    print(f'ACCURACY SCORE:{accuracy_score(y_test,predictions)}')
    print(f'CONFUSION MATRIX:{confusion_matrix(y_test,predictions)}')
    print(f'CLASSIFIFCATIONREPORT:{classification_report(y_test,predictions)}')
    print('\n')
```

```python

from sklearn.model_selection import GridSearchCV
```

```python
Ranf =  RandomForestClassifier()
```

```python
param_grid = {
    'n_estimators': [ 150,250,350,450],
    'max_depth': [5, 10, 15, 20, 25],
    'min_samples_split': [2,5,10],
    'min_samples_leaf':[1,2,4]

}
```

```python
grid = GridSearchCV(Ranf,param_grid,cv=3,n_jobs=1,verbose=2,scoring= 'accuracy')
grid.fit(X_train,y_train)
```

```python
best_param = grid.best_params_
print(f'Best paratmeter is : {best_param}')
```

```python
best_model = RandomForestClassifier(**best_param)
best_model.fit(X_train,y_train)

prediction = best_model.predict(X_test)
```

```python
X_train.columns 
```

```python
print('Evaluation')

print(f'ACCURACY SCORE:{accuracy_score(y_test,prediction)}')

print(f'CONFUSION MATRIX:{confusion_matrix(y_test,prediction)}')

print(f'CLASSIFIFCATIONREPORT:{classification_report(y_test,prediction)}')

print('\n')
```

```python
sns.heatmap(data = confusion_matrix(y_test,prediction),annot= True)
```

```python
new_loan  = [[1,0,1,3000,500,200,1,30]]

prediction = best_model.predict(new_loan)

print(prediction)

if prediction[0] == 1:
    print('Loan is approved')
else:
    print('Loan is not approved')
    
```
