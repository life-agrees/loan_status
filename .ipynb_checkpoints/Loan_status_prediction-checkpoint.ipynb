```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
loan["gender"] = label_encoder.fit_transform(loan["Gender"])
loan["married"] = label_encoder.fit_transform(loan["Married"])
loan["education"] = label_encoder.fit_transform(loan["Education"])
loan["self_employed"] = label_encoder.fit_transform(loan["Self_Employed"])
loan["property_area"] = label_encoder.fit_transform(loan["Property_Area"])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Extracting my X feature
```

```python
X = loan.drop("Loan_Status",axis = 1)
y = loan["Loan_Status"]
```

```python
#let's create an instance 
```

```python
scale = StandardScaler()
scaler = scale.fit_transform(X)
scaled = pd.DataFrame(scaler,columns = X.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_shape = X_train,X_test
y_shape = y_train,y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logistic = LogisticRegression()
```

```python
Logistic.fit(X_train  , y_train)
```

```python
make_prediction  = Logistic.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print("classification_report('_'):",classification_report(y_test,make_prediction))
print("confusion_matrix('_'):",confusion_matrix(y_test,make_prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
d_tree = DecisionTreeClassifier()
rand = RandomForestClassifier(n_estimators=150)
support = SVC()
```

```python
d_tree.fit(X_train,y_train)
rand.fit(X_train,y_train)
support.fit(X_train,y_train)
```

```python
d_pred = d_tree.predict(X_test)
r_pred = rand.predict(X_test)
s_pred = support.predict(X_test)
```

```python
#print("ClASSIFICATION_REPORT:",classification_report(y_test,d_pred),'\n',classification_report(y_test,r_pred),'\n',
     ##classification_report(y_test,s_pred))

#print("CONFUSION_MATRIX:",confusion_matrix(y_test,d_pred),'\n',confusion_matrix(y_test,r_pred),'\n',
     #confusion_matrix(y_test,s_pred))
```

```python

print("CLASSIFICATION_REPORT:")
print("Decision Tree:")
print(classification_report(y_test, d_pred))
print("\nRandom Forest:")
print(classification_report(y_test, r_pred))
print("\nSVC:")
print(classification_report(y_test, s_pred))

print("\nCONFUSION_MATRIX:")
print("Decision Tree:")
print(confusion_matrix(y_test, d_pred))
print("\nRandom Forest:")
print(confusion_matrix(y_test, r_pred))
print("\nSVC:")
print(confusion_matrix(y_test, s_pred))
```
